```python
from selenium.webdriver.common.by import By
import time

def get_listing_urls(driver, city):

    search_url = f"https://www.airbnb.com/s/{city}/homes"

    driver.get(search_url)

    time.sleep(10)

    links = driver.find_elements(By.TAG_NAME, "a")

    listing_urls = []

    for link in links:

        href = link.get_attribute("href")

        if href and "/rooms/" in href:
            listing_urls.append(href)

    return list(set(listing_urls))
```

```python
def scrape_listing(driver, url, city):

    try:

        driver.get(url)

        time.sleep(5)

        property_name = driver.find_element(
            By.TAG_NAME,
            "h1"
        ).text

        property_type = driver.find_element(
            By.TAG_NAME,
            "h2"
        ).text

        page_text = driver.page_source.lower()

        superhost = "superhost" in page_text

        return {
            "city": city,
            "property_name": property_name,
            "property_type": property_type,
            "superhost": superhost,
            "url": url
        }

    except Exception as e:

        print(f"Error: {url}")

        return None
```

```python
from selenium.webdriver.common.by import By
import time

def get_listing_urls(driver, city, target=50):

    search_url = f"https://www.airbnb.com/s/{city}/homes"

    driver.get(search_url)

    time.sleep(8)

    urls = set()

    last_count = 0

    for _ in range(20):

        links = driver.find_elements(By.TAG_NAME, "a")

        for link in links:

            href = link.get_attribute("href")

            if href and "/rooms/" in href:
                urls.add(href)

        print(f"Collected: {len(urls)}")

        if len(urls) >= target:
            break

        driver.execute_script(
            "window.scrollTo(0, document.body.scrollHeight);"
        )

        time.sleep(4)

        if len(urls) == last_count:
            break

        last_count = len(urls)

    return list(urls)[:target]
```

```python
from selenium.webdriver.common.by import By
import time

def get_listing_urls(driver, city):

    search_url = f"https://www.airbnb.com/s/{city}/homes"

    driver.get(search_url)

    time.sleep(10)

    urls = set()

    links = driver.find_elements(By.TAG_NAME, "a")

    for link in links:

        href = link.get_attribute("href")

        if href and "/rooms/" in href:
            urls.add(href)

    return list(urls)
```

```python
import re
import time

def scrape_listing(driver, url, city):

    try:

        driver.get(url)

        time.sleep(8)

        body = driver.find_element(
            By.TAG_NAME,
            "body"
        ).text

        property_name = driver.find_element(
            By.TAG_NAME,
            "h1"
        ).text

        h2_tags = driver.find_elements(
            By.TAG_NAME,
            "h2"
        )

        property_type = (
            h2_tags[0].text
            if len(h2_tags) > 0
            else None
        )

        superhost = (
            "superhost" in body.lower()
        )

        guest_favorite = (
            "guest favorite" in body.lower()
        )

        ratings = re.findall(
            r"\b[0-5]\.\d+\b",
            body
        )

        rating = (
            float(ratings[0])
            if ratings
            else None
        )

        review_match = re.search(
            r"(\d+)\s+reviews",
            body,
            re.IGNORECASE
        )

        reviews_count = (
            int(review_match.group(1))
            if review_match
            else None
        )

        return {
            "city": city,
            "property_name": property_name,
            "property_type": property_type,
            "rating": rating,
            "reviews_count": reviews_count,
            "superhost": superhost,
            "guest_favorite": guest_favorite,
            "url": url
        }

    except Exception as e:

        print(f"Error: {url}")

        return None
```

```python
import pandas as pd
import random

cities = [
    'amsterdam',
    'athens',
    'barcelona',
    'berlin',
    'budapest',
    'lisbon',
    'paris',
    'rome',
    'vienna',
    'london'
]

all_data = []

for city in cities:

    print("=" * 60)
    print(f"Scraping {city}")

    urls = get_listing_urls(
        driver,
        city
    )

    print(f"Found {len(urls)} Listings")

    # ناخد أول 30 عقار
    urls = urls[:30]

    for i, url in enumerate(urls, start=1):

        print(f"{city} | Listing {i}")

        row = scrape_listing(
            driver,
            url,
            city
        )

        if row:
            all_data.append(row)

        time.sleep(
            random.uniform(2, 5)
        )
```

```python
scraped_df = pd.DataFrame(all_data)
scraped_df
```
