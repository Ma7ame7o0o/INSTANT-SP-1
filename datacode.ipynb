**Setup & Imports**

```python
import pandas as pd
from pathlib import Path

# Configuration
DATA_FOLDER = Path("data")
OUTPUT_FILE = Path("airbnb_combined.csv")
```

**Processing & Merging Data**

```python
csv_files = sorted(DATA_FOLDER.glob("*.csv"))

if not csv_files:
    print(f"Error: No CSV files found in {DATA_FOLDER.resolve()}")
else:
    all_dfs = []
    
    for file_path in csv_files:
        filename = file_path.stem.lower()
        
        # Extract metadata from filename
        if "weekdays" in filename:
            day_type = "weekday"
            city = filename.replace("_weekdays", "")
        elif "weekends" in filename:
            day_type = "weekend"
            city = filename.replace("_weekends", "")
        else:
            continue
            
        df = pd.read_csv(file_path)
        
        # Clean Unnamed columns
        df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
        
        # Add metadata columns
        df["city"] = city.capitalize()
        df["day_type"] = day_type
        
        all_dfs.append(df)
        print(f"Processed: {file_path.name} | Rows: {len(df)}")

    # Combine all dataframes
    combined_df = pd.concat(all_dfs, ignore_index=True, sort=False)
    print(f"\nTotal rows: {len(combined_df)}")
```

```python
# Check row distribution
print("Row count by City and Day Type:")
print(combined_df.groupby(["city", "day_type"]).size())

# Export
combined_df.to_csv(OUTPUT_FILE, index=False)
print(f"\nSaved to: {OUTPUT_FILE.resolve()}")
```
